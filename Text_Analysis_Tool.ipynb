# Algorithm for Visualization

```python
nlp = spacy.load("en_core_web_sm")


def topic_header(bullet):

    WINDOW = 5
    doc = nlp(bullet)

    span = None
    for nc in doc.noun_chunks:
        if any(t.pos_.startswith("N") or t.pos_ == "PROPN" for t in nc):
            span = nc
            break
    if span is None:
        for t in doc:
            if t.pos_ in {"NOUN", "PROPN"}:
                span = doc[t.i:t.i+1]
                break
    if span is None:
        return bullet

    start, end = span.start, span.end
    left_idx  = max(0, start - WINDOW)
    right_idx = min(len(doc), end + WINDOW)

    left  = " ".join(tok.text for tok in doc[left_idx:start])
    mid   = span.text
    right = " ".join(tok.text for tok in doc[end:right_idx])

    prefix = "..." if left_idx > 0 else ""
    suffix = "..." if right_idx < len(doc) else ""

    snippet = f"{left} [{mid}] {right}".strip()
    if prefix:
        snippet = f"{prefix} {snippet}"
    if suffix:
        snippet = f"{snippet} {suffix}"
    return snippet


combined = []
for b in remote_pros_A: combined.append(("pro", b, "A", topic_header(b)))
for b in remote_cons_A: combined.append(("con", b, "A", topic_header(b)))
for b in remote_pros_B: combined.append(("pro", b, "B", topic_header(b)))
for b in remote_cons_B: combined.append(("con", b, "B", topic_header(b)))


dot = Graph("comparison_panels")
dot.attr(rankdir="LR", fontsize='20')

clusters = [("pro", "PROS"), ("con", "CONS")]
panel_bg = {"pro": "#99ff99", "con": "#ff9b99"}
anchor_ids = {}

for key, title in clusters:
    with dot.subgraph(name=f"cluster_{key}") as c:
        c.attr(label=title, style='rounded,filled', bgcolor=panel_bg[key])
        anchor_id = f"{key}_anchor"
        anchor_ids[key] = anchor_id
        c.node(anchor_id, label="", shape="point", width="0.01", height="0.01", style="invis")

        items = [(i, stance, text, src, topic) for i, (stance, text, src, topic) in enumerate(combined) if stance == key]

        for i, _, text, src, topic in items:
            node_id = f"{key}_{src}_{i}"
            shape = "box"
            c.node(node_id, label=f"{topic}\n{src}", shape=shape, style="rounded,filled", fillcolor="#fcfcfc")

dot.edge(anchor_ids["pro"], anchor_ids["con"], style="invis")

dot.format = "png"
dot.graph_attr.update(dpi="300", bgcolor="transparent")  # sharper PNG, transparent bg (optional)

outpath = dot.render(filename="figs/visualization", cleanup=True)  # writes figs/visualization.png
files.download("figs/visualization.png")
```
